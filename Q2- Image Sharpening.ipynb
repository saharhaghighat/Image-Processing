```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread('2-1.png')
img2 = cv.imread('2-2.png')

gray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

alpha1=1
alpha=1.5
alpha5 = 5

```

# Laplacian Sharpening

```python



laplacian1=cv.Laplacian(gray1,cv.CV_64F)
laplacian1 = cv.convertScaleAbs(laplacian1)

sharp_laplacian_moon1 = cv.addWeighted(gray1, 1, laplacian1, alpha1, 0)
sharp_laplacian_moon2=cv.addWeighted(gray1, 1, laplacian1, alpha, 0)
sharp_laplacian_moon3=cv.addWeighted(gray1, 1, laplacian1, alpha5, 0)


#
laplacian2=cv.Laplacian(gray2,cv.CV_64F)
laplacian2 = cv.convertScaleAbs(laplacian2)
sharp_laplacian_F1 = cv.addWeighted(gray2, 1, laplacian2, alpha1, 0)
sharp_laplacian_F2=cv.addWeighted(gray2, 1, laplacian2, alpha, 0)
sharp_laplacian_F3=cv.addWeighted(gray2, 1, laplacian2, alpha5, 0)

plt.figure(figsize=(24, 12))
plt.subplot(2, 5, 1)
plt.title('Original Image 1')
plt.imshow(img1)
plt.axis('off')


plt.subplot(2, 5, 2)
plt.title('Laplacian Image 1')
plt.imshow(laplacian1, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 3)
plt.title('Sharpened Laplacian Image 1 with alpha 1')
plt.imshow(sharp_laplacian_moon1,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 4)
plt.title('Sharpened Laplacian Image 1 with alpha 1.5')
plt.imshow(sharp_laplacian_moon2,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 5)
plt.title('Sharpened Laplacian Image 1 with alpha 5')
plt.imshow(sharp_laplacian_moon3,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 6)
plt.title('Original Image 2')
plt.imshow(img2)
plt.axis('off')

plt.subplot(2, 5, 7)
plt.title('Laplacian Image 2')
plt.imshow(laplacian2, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 8)
plt.title('Sharpened Laplacian Image 2 with alpha 1')
plt.imshow(sharp_laplacian_F1, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 9)
plt.title('Sharpened Laplacian Image 2 with alpha 1.5')
plt.imshow(sharp_laplacian_F2, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 10)
plt.title('Sharpened Laplacian Image 2 with alpha 5')
plt.imshow(sharp_laplacian_F3, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Gaussian sharpening

```python
blurred1 = cv.GaussianBlur(gray1, (5, 5), 0)
sharpened_gaussian_moon1 = cv.addWeighted(gray1, alpha1, blurred1, -0.5, 0)
sharpened_gaussian_moon2 = cv.addWeighted(gray1, alpha, blurred1, -0.5, 0)
sharpened_gaussian_moon3 = cv.addWeighted(gray1, alpha5, blurred1, -0.5, 0)



blurred2 = cv.GaussianBlur(gray2, (5, 5), 0)
sharpened_gaussian_F1 = cv.addWeighted(gray2, alpha1, blurred2, -0.5, 0)
sharpened_gaussian_F2 = cv.addWeighted(gray2, alpha, blurred2, -0.5, 0)
sharpened_gaussian_F3 = cv.addWeighted(gray2, alpha5, blurred2, -0.5, 0)




plt.figure(figsize=(24, 12))
plt.subplot(2, 5, 1)
plt.title('Original Image 1')
plt.imshow(img1)
plt.axis('off')


plt.subplot(2, 5, 2)
plt.title('Blurred Image 1')
plt.imshow(blurred1, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 3)
plt.title('Sharpened Gaussian Image 1 with alpha 1')
plt.imshow(sharpened_gaussian_moon1,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 4)
plt.title('Sharpened Gaussian Image 1 with alpha 1.5')
plt.imshow(sharpened_gaussian_moon2,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 5)
plt.title('Sharpened Gaussian Image 1 with alpha 5')
plt.imshow(sharpened_gaussian_moon3,cmap='gray' )
plt.axis('off')

plt.subplot(2, 5, 6)
plt.title('Original Image 2')
plt.imshow(img2)
plt.axis('off')

plt.subplot(2, 5, 7)
plt.title('Blurred Image 2')
plt.imshow(blurred2, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 8)
plt.title('Sharpened Gaussian Image 2 with alpha 1')
plt.imshow(sharpened_gaussian_F1, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 9)
plt.title('Sharpened Gaussian Image 2 with alpha 1.5')
plt.imshow(sharpened_gaussian_F2, cmap='gray')
plt.axis('off')

plt.subplot(2, 5, 10)
plt.title('Sharpened Gaussian Image 2 with alpha 5')
plt.imshow(sharpened_gaussian_F3, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
