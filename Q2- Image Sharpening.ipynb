```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread('2-1.png')
img2 = cv.imread('2-2.png')

gray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
```

# Laplacian Sharpening

```python
laplacian1=cv.Laplacian(gray1,cv.CV_64F)
laplacian1 = cv.convertScaleAbs(laplacian1)
sharp_laplacian1 = gray1 + laplacian1
sharp_laplacian1 = np.clip(sharp_laplacian1, 0, 255).astype(np.uint8)

laplacian2=cv.Laplacian(gray2,cv.CV_64F)
laplacian2 = cv.convertScaleAbs(laplacian2)
sharp_lapacian2 = gray2 + laplacian2
sharp_laplacian2 = np.clip(sharp_lapacian2, 0, 255).astype(np.uint8)


plt.figure(figsize=(18, 6))
plt.subplot(2, 3, 1)
plt.title('Original Image 1')
plt.imshow(img1)
plt.axis('off')


plt.subplot(2, 3, 2)
plt.title('Laplacian Image 1')
plt.imshow(laplacian1, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title('Sharpened Laplacian Image 1')
plt.imshow(sharp_laplacian1,cmap='gray' )
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title('Original Image 2')
plt.imshow(img2)
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title('Laplacian Image 2')
plt.imshow(laplacian2, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.title('Sharpened Laplacian Image 2')
plt.imshow(sharp_laplacian2, cmap='gray')
plt.axis('off')


plt.tight_layout()
plt.show()
```

# Combining

```python
# Creating a combined image for visualization
sharp_img_combined1 = np.zeros_like(img1, dtype=np.uint8)
sharp_img_combined1[:,:,0] = np.clip(img1[:,:,0] + sharp_laplacian1, 0, 255)
sharp_img_combined1[:,:,1] = np.clip(img1[:,:,1] + sharp_laplacian1, 0, 255)
sharp_img_combined1[:,:,2] = np.clip(img1[:,:,2] + sharp_laplacian1, 0, 255)

sharp_img_combined2 = np.zeros_like(img2, dtype=np.uint8)
sharp_img_combined2[:,:,0] = np.clip(img2[:,:,0] + sharp_laplacian2, 0, 255)
sharp_img_combined2[:,:,1] = np.clip(img2[:,:,1] + sharp_laplacian2, 0, 255)
sharp_img_combined2[:,:,2] = np.clip(img2[:,:,2] + sharp_laplacian2, 0, 255)

# Plotting the results
plt.figure(figsize=(18, 6))

plt.subplot(2, 2, 1)
plt.title('Original Image 1')
plt.imshow(img1)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Combiened 1')
plt.imshow(cv.cvtColor(sharp_img_combined1, cv.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Original Image 2')
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Combiened 2')
plt.imshow(cv.cvtColor(sharp_img_combined2, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.tight_layout()
plt.show()
```
