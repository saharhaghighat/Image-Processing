```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage 
```

```python
#Reading Image
img = cv.imread('1-1.webp')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gaussian = cv.GaussianBlur(gray,(3,3),0)
```

# 1.1 Roberts

```python
img2 = cv.imread("1-1.webp",0).astype('float64') 
img2/=255.0

roberts_cross_v = np.array( [[1, 0 ], 
                             [0,-1 ]] ) 
  
roberts_cross_h = np.array( [[ 0, 1 ], 
                             [ -1, 0 ]] ) 

vertical = ndimage.convolve(img2,roberts_cross_v ) 
horizontal = ndimage.convolve( img2, roberts_cross_h ) 
  
edged_img = np.sqrt( np.square(horizontal) + np.square(vertical)) 
edged_img*=255
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Roberts Edge Detection')
plt.imshow(edged_img)
plt.axis('off')
```

# 1.1 ‫‪Prewitt‬‬

```python
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv.filter2D(img_gaussian, -1, kernely)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Prewitt Edge Detection with kernel x')
plt.imshow(img_prewittx)
plt.axis('off')
plt.subplot(1, 3, 1)
plt.title('Prewitt Edge Detection with kernel y')
plt.imshow(img_prewitty)
plt.axis('off')
```

# 1.1 Sobel 

```python
img_sobelx = cv.Sobel(img_gaussian,cv.CV_8U,1,0,ksize=5)
img_sobely = cv.Sobel(img_gaussian,cv.CV_8U,0,1,ksize=5)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('img_sobelx Edge Detection ')
plt.imshow(img_sobelx)
plt.axis('off')
plt.subplot(1, 3, 1)
plt.title('img_sobely Edge Detection')
plt.imshow(img_sobely)
plt.axis('off')
```

# 1.1 Laplacian

```python
laplacian = cv.Laplacian(img, cv.CV_64F)
laplacian = cv.convertScaleAbs(laplacian)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Laplacian Edge Detection')
plt.imshow(laplacian)
plt.axis('off')
```
